```python
import fitz
import re  # Import the regular expression library

def extract_arabic_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

def clean_arabic_text(text):
    """Cleans Arabic text by replacing specific patterns and removing harakat."""

    # Replace "اال" with "الا"
    text = text.replace("اال", "الا")
    text = text.replace("األ", "الأ")
    text = text.replace("اإل", "الإ")
    text = text.replace("اآل", "الآ")

    # Remove harakat followed by space
    text = re.sub(r"[\u064B-\u0652]\s", "", text)

    # Remove all harakat
    text = re.sub(r"[\u064B-\u0652]", "", text)

    return text

def split_into_sentences(text):
    """Splits Arabic text into sentences, handling punctuation and abbreviations."""

    # Define a regular expression pattern to match sentence endings
    pattern = r"(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s"

    # Split the text based on the pattern
    sentences = re.split(pattern, text)

    return sentences
```

```python
pdf_path = "/content/original.pdf"
arabic_text = extract_arabic_text_from_pdf(pdf_path)
cleaned_text = clean_arabic_text(arabic_text)
sentences = split_into_sentences(cleaned_text)
```

```python
import torch
import pandas as pd
from sentence_transformers import SentenceTransformer

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = SentenceTransformer('intfloat/multilingual-e5-large', device=device)

embeddings = model.encode(sentences)

data = {'sentence': sentences, 'embedding': embeddings.tolist()}
df = pd.DataFrame(data)
df.to_csv("document_dataset", index=False)
```

# Get answer chunk

```python
def get_common_words(text1, text2):
    """Counts the number of common words between two texts."""
    words1 = set(text1.split())
    words2 = set(text2.split())
    return len(words1.intersection(words2))

sentence_indices = []
for _, row in df.iterrows():
    if type(row["context"]) != str:
      best_sentence_index = -1
      sentence_indices.append(best_sentence_index)
      continue
    cleaned_context = clean_arabic_text(row["context"])
    max_common_words = 0
    best_sentence_index = -1
    for i, sentence in enumerate(sentences):
        common_words = get_common_words(cleaned_context, sentence)
        if common_words > max_common_words:
            max_common_words = common_words
            best_sentence_index = i
    sentence_indices.append(best_sentence_index)
df["chunk"] = sentence_indices
df.to_csv("edited.csv")
```

```python
import pandas as pd
from sentence_transformers import util

# Function to calculate cosine similarities and retrieve sentences
def retrieve_top_sentences(question_embedding, all_sentences_embeddings, gold_idx, k=20):
    similarities = util.cos_sim(question_embedding, all_sentences_embeddings)
    rank = (similarities[0] >= similarities[0, gold_idx]).sum()
    top_k_indices = similarities[0].argsort(descending=True)[:k]
    return df_documents.iloc[top_k_indices], rank.item()
```

```python
import numpy as np
import ast
from tqdm import tqdm


# Load data
df_questions = pd.read_csv("dataset_with_embeddings.csv")
df_documents = pd.read_csv("document_dataset3.csv")

def str_to_array(s):
    # Using ast.literal_eval to safely evaluate the string
    return np.array(ast.literal_eval(s))

df_questions['question_formal_embedding'] = df_questions['question_formal_embedding'].apply(str_to_array)
df_questions['question_conversational_embedding'] = df_questions['question_conversational_embedding'].apply(str_to_array)
df_documents['refined_embedding'] = df_documents['refined_embedding'].apply(str_to_array)

# Iterate over questions
c1 = 0
a1 = 0
b1 = 0

c2 = 0
a2 = 0
b2 = 0
for idx, row in tqdm(df_questions.iterrows(), total=df_questions.shape[0], desc="Processing Questions"):
    for question_type in ["formal", "conversational"]:
        question = row[f"question_{question_type}"]
        embedding = row[f"question_{question_type}_embedding"]
        if "," in row['chunck']:
          true_idx = int(row['chunck'][:2])
        else:
          true_idx = int(row['chunck'])
        # Retrieve top 20 sentences
        top_20, rank = retrieve_top_sentences(embedding, df_documents['refined_embedding'].tolist(), true_idx, 20)
        if question_type == "formal":
          c1 += 1
          if rank <= 10:
            a1 += 1
          elif rank <= 20:
            b1 +=1
        else:
          c2 += 1
          if rank <= 10:
            a2 += 1
          elif rank <= 20:
            b2 +=1
```
